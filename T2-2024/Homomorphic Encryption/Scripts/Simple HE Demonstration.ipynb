### Simple Homographic Encryption Demo:
- **Encryption:** The numbers 1 and 2 are individually encrypted using the public key.
- **Homomorphic Addition:** The encrypted numbers are added together, producing an encrypted result without ever needing to decrypt them first.
- **Decryption:** The sum of the encrypted numbers is decrypted using the private key to reveal the result.

```python
from phe import paillier

# Generate public and private keys
public_key, private_key = paillier.generate_paillier_keypair()

# Encrypt two numbers: 1 and 2
encrypted_1 = public_key.encrypt(1)
encrypted_2 = public_key.encrypt(2)

# Perform homomorphic addition (encrypted_1 + encrypted_2)
encrypted_sum = encrypted_1 + encrypted_2

# Decrypt the result
decrypted_sum = private_key.decrypt(encrypted_sum)

# Print the results
print(f"Encrypted 1: {encrypted_1.ciphertext()}")
print(f"Encrypted 2: {encrypted_2.ciphertext()}")
print(f"Encrypted sum (enc(1) + enc(2)): {encrypted_sum.ciphertext()}")
print(f"Decrypted sum: {decrypted_sum}")
```
